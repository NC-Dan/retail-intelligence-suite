# E-commerce Customer Analytics: From Raw Data to Executive Dashboard

##  Project Overview
This end-to-end analytics project analyzes UK-based online retail transaction data from December 2010 to December 2011. The company specializes in unique all-occasion gifts with many wholesale customers. The project demonstrates the complete data science workflow from raw data processing to executive-ready business intelligence.

##  Business Objectives
- **Customer Segmentation**: Identify high-value customer segments using RFM analysis
- **Churn**: Predict which customers are likely to stop purchasing
- **Revenue Optimization**: Provide actionable insights for business growth
- **Executive Reporting**: Create interactive dashboards for stakeholder decision-making

##  Technical Stack
- **Python**: Data processing, analysis, and modeling
- **Pandas & NumPy**: Data manipulation and numerical computing
- **Matplotlib & Seaborn**: Data visualization
- **Power BI**: Interactive dashboard creation
- **Jupyter Notebooks**: Development environment

##  Key Insights & Results
- **Customer Segments Identified**: 9 distinct customer segments with varying value profiles
- **Revenue Impact**: Potential 15-25% revenue increase through targeted retention strategies
- **Seasonal Patterns**: Clear seasonal trends affecting sales performance

##  Business Recommendations
1. **Focus on High-Value Segments**: Prioritize marketing spend on Champions and Loyal customers
2. **Retention Campaigns**: Target at-risk customers identified.
3. **Seasonal Planning**: Optimize inventory and marketing for identified seasonal patterns
4. **Wholesale Strategy**: Develop specific strategies for wholesale vs. retail customers

- **Geography**: United Kingdom
- **Business Type**: Non-store online retail (Gift retailer)
- **Customer Base**: Mix of individual customers and wholesalers
